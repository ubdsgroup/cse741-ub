```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
import warnings
warnings.filterwarnings("ignore")
```

```python
# Create a random dataset
rng = np.random.RandomState(1)
X = np.sort(200 * rng.rand(600, 1) - 100, axis=0)
y = np.array([np.pi * np.sin(X).ravel(), np.pi * np.cos(X).ravel()]).T
y += (0.5 - rng.rand(*y.shape))

X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=400, test_size=200, random_state=4)
```

```python
max_depth = 30
regr_multirf = MultiOutputRegressor(RandomForestRegressor(n_estimators=100,
                                                          max_depth=max_depth,
                                                          random_state=0))
regr_multirf.fit(X_train, y_train)
# Predict on new data
y_multirf = regr_multirf.predict(X_test)
```

```python
# Plot the results
plt.figure()
s = 50
a = 0.4
plt.scatter(y_test[:, 0], y_test[:, 1], edgecolor='k',
            c="navy", s=s, marker="s", alpha=a, label="Data")
plt.scatter(y_multirf[:, 0], y_multirf[:, 1], edgecolor='k',
            c="cornflowerblue", s=s, alpha=a,
            label="Multi RF score=%.2f" % regr_multirf.score(X_test, y_test))
plt.xlim([-6, 6])
plt.ylim([-6, 6])
plt.show()
```

## Discover conservation laws expressed as linear equations

What are conservation laws?
- Example: Law of Conservation of Energy

Mathematically:
<img src='conservationlaw.png' width="400"/>

$u(x)$ and $f(x)$ are two functions connected via a **parametric linear operator**:
$$
f(x) = \mathcal{L}^\phi_x(u(x))
$$

## Linear Operator
- *Operator* - another term for a function, mapping or transformation
- e.g., $$g(x) = \int_a^b k(x,t)f(t)dt$$
  gives a new function $g()$ from the function $f()$ by applying a kernel ($k(,)$)
- Can be thought of as a generalization of a matrix-vector product
- An operator $A$ is linear if: $$A(av_1 + bv_2) = aAv_1 + bAv_2$$

## Traditional ML
- Given noisy observations $\left\{{\bf X}_u,{\bf y}_u\right\}$, learn $u(x)$
  * Can use GP for this
- Same thing for $f(x)$

Here it is a little different:
> We need to learn $\phi$ - governing equation that describes the data
  - *inverse problems*

## Method

Assume that $u(x)$ is a GP with 0 mean and covariance function $k_{uu}(x,x';\theta)$ where $\theta$ are the hyper-parameters of the kernel function $k_{uu}()$. 

$$
u(x) \sim \mathcal{GP}(0,k_{uu}(x,x';\theta))
$$
A *key* observation is that a linear transformation of a GP will also be GP. Thus:
$$
\mathcal{L}_x^\phi u(x) = f(x) \sim \mathcal{GP}(0,k_{ff}(x,x';\theta,\phi))
$$

where the new kernel function $k_{ff}$ is given by:
$$
k_{ff}(x,x';\theta,\phi) = \mathcal{L}_x^\phi\mathcal{L}_{x'}^\phi k_{uu}(x,x';\theta)
$$
> Note how the parameter $\phi$ is turned into hyper-parameter of $k_{ff}$

## The kernel
Using the widely used squared exponential covariance function
$$
k_{uu}(x,x';\theta) = \sigma^2_u \exp\left(-\frac{1}{2}\sum_{d=1}^Dw_d(x_d - x'_d)^2\right)
$$
> Allows for **automatic relevance determination** (ARD)

## Training
- Need to learn the hyper-parameters in $\theta$ and the parameters $\phi$ of the linear operator
- GP gives a nice optimization formulation in the form of the **negative log marginal likelihood**

$$
\mathcal{NLML}(\phi,\theta,\sigma^2_{n_u},\sigma^2_{n_f}) = -\log{p({\bf y}\vert \phi,\theta,\sigma^2_{n_u},\sigma^2_{n_f} )}
$$
- where ${\bf y} = \left[\begin{array}{c}{\bf y}_u\\{\bf y}_f\end{array}\right]$

## Prediction
- Fairly straightforward extension of standard GP.
- Allows getting both $f(x)$ and $u(x)$ for a new $x$

### Synthetic Data
* Assume some $u(x)$ and $f(x)$
* The linear operator is:
  $$
  \mathcal{L}^{(\alpha,\beta)}_x u(x) := \frac{d}{dx}u(x) + \alpha u(x) + \beta\int_0^x u(\xi)d\xi = f(x)
  $$
  > An integro-differential equation
  
* By using noisy samples $\{{\bf x}_u,{\bf y}_u\}$ and $\{{\bf x}_f,{\bf y}_f\}$ for training, one can learn the parameters, $\alpha$ and $\beta$.

## Heat Equation

$$
\mathcal{L}^{\alpha}_{(t,x)} u(t,x) := \frac{\partial}{\partial t}u(t,x) - \alpha \frac{\partial^2}{\partial x^2}u(t,x) = f(t,x)
$$

- For some values of $\alpha$, $u(t,x)$ and $f(t,x)$ are given

## Drosophila melanogaster gap gene dynamics 
<img src='https://upload.wikimedia.org/wikipedia/commons/9/95/Drosophila_melanogaster_Proboscis.jpg' width="400"/>

* Understanding the gap gene dynamics of a protein (denoted by $a$)
* Reaction-diffusion PDE
$$
\mathcal{L}^{(\lambda^a,D^a)}_{(t,x)} u^a(t,x) := \frac{\partial}{\partial t}u^a(t,x) + \lambda^a u^a(t,x) -D^a\frac{\partial^2}{\partial x^2} u^a(t,x)= f^a(t,x)
$$
